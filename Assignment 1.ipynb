```python
# 2.
# The given code is about to find the root of a given equation with tolerance and 
# using Absolute Error as stopping criteria

import numpy as np
import pandas as pd

def f(x):
    return np.cos(x) - x * np.exp(x)

def bisection(a, b,e):
    i = 1
    c_old = 0
    table = []
    condition = True
    while condition:
        c = (a + b) / 2

        abs_error = abs(c - a)
        table.append([i, a ,b , c , abs_error])
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c

        if abs_error < e:
            condition = False

        c_old = c
        i += 1
    df = pd.DataFrame(table , columns= (['No. of Iteration required','a','b','Aprroximate value of root','Absolute error']))  
    pd.set_option('display.float_format', '{:.7f}'.format)
    return df, c

a = float(input("Enter the first guess value: "))
b = float(input("Enter the second guess value: "))
e = float(input("Tolerance is: "))

if f(a) * f(b) > 0:
    print("No root in this interval")
else:
    df, c = bisection(a, b, e)
    print(df.to_string(index = False))
    print("Approximate final root is: ",round(c, 7))
```
