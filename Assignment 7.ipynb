```python
# 5
import sympy as sy
import pandas as pd

w, x, y, z = sy.symbols('w x y z')
l_mat = sy.zeros(4, 4)
u_mat = sy.zeros(4, 4)
d_mat = sy.zeros(4, 4)
table = []
lhs = sy.Matrix([[4,1,-1,1],
                 [1,4,-1,-1],
                 [-1,-1,5,1],
                [1,-1,1,3]])
rhs = sy.Matrix([[-2], [-1], [0],[1]])
vector = sy.Matrix([[w], [x], [y], [z]])

for i in range(4):
    for j in range(4):
        if i == j:
            d_mat[i, j] = lhs[i, j]
        elif i > j:
            l_mat[i, j] = lhs[i, j]
        elif i < j:
            u_mat[i, j] = lhs[i, j]

if sy.det(lhs) != 0:
    inv_d = d_mat.inv()
else:
    raise ValueError("Matrix not invertible")

exact_solution = lhs.inv() @ rhs
print("\nExact solution:")
sy.pprint(exact_solution.evalf(5))

tol = 1e-3
def gauss_jacobi(n, vector):
    vector = sy.Matrix([0,0,0,0])
    table = []
    for i in range(1,n+1):
        next_vector = -(inv_d)@(l_mat + u_mat)@vector + inv_d@rhs
        error = (abs(next_vector - exact_solution)).evalf(3)
        table.append([i , next_vector.evalf(3), error.evalf(3)])
        if all(e < tol for e in error):
            df1 = pd.DataFrame(table, columns=(['Iteration','Approx Solution', 'Error']))
            return df1, next_vector
        vector = next_vector
    df1 = pd.DataFrame(table, columns=(['Iteration','Approx Solution', 'Error']))
    return df1, vector 

df1, vector = gauss_jacobi(10, vector)
print("\nJacobi Method with 10 Iterations:")
print(df1.to_string(index=False))

df1, vector = gauss_jacobi(20, vector)
print("\nJacobi Method with 20 Iterations:")
print(df1.to_string(index=False))
print(f'After acheiving tolerance loop automatically stop at 13 iteration ')
```

```python
#  6
import sympy as sy
import pandas as pd

w, x, y, z = sy.symbols('w x y z')

lhs = sy.Matrix([[4,1,-1,1],
                 [1,4,-1,-1],
                 [-1,-1,5,1],
                [1,-1,1,3]])
rhs = sy.Matrix([[-2], [-1], [0],[1]])
vector = sy.Matrix([[w], [x], [y], [z]])

d_mat = sy.zeros(4, 4)
l_mat = sy.zeros(4, 4)
u_mat = sy.zeros(4, 4)
table = []
for i in range(4):
    for j in range(4):
        if i == j:
            d_mat[i, j] = lhs[i, j]
        elif i > j:
            l_mat[i, j] = lhs[i, j]
        elif i < j:
            u_mat[i, j] = lhs[i, j]

if sy.det(lhs) != 0:
    inv_d = d_mat.inv()
else:
    raise ValueError("Matrix not invertible")

exact_solution = lhs.inv() @ rhs
print("\nExact solution:")
sy.pprint(exact_solution.evalf(5))

tol = 1e-3

def gauss_seidel(n, vector ):
    vector = sy.Matrix([0, 0, 0, 0])
    ld_mat = d_mat + l_mat
    inv_ld = ld_mat.inv()
    for i in range(1, n + 1):
        next_vector = -inv_ld @ u_mat @ vector + inv_ld @ rhs
        error = (abs(next_vector - exact_solution)).evalf(3)
        table.append([i ,next_vector.evalf(3), (error).evalf(3)])
        if all(e < tol for e in error):
            df2 = pd.DataFrame(table, columns=(['Iteration','Approx Solution', 'Error']))
            return df2, next_vector
        vector = next_vector
    df2 = pd.DataFrame(table, columns=(['Iteration','Approx Solution', 'Error']))
    return df2, vector

df1 , vector = gauss_jacobi( 20, sy.Matrix([0,0,0,0]))
df2 , vector = gauss_seidel(20 , sy.Matrix([0,0,0,0]))


print("\nComparison of Jacobi and Gauss-Seidel Methods:")
print("Jacobi Method (Last Iteration):")
print(df1.to_string(index=False))
print(f'After acheiving tolerance loop automatically stop at 13 iteration ')
print("\nGauss-Seidel Method (Last Iteration):")
print(df2.to_string(index=False))
print(f'After acheiving tolerance loop automatically stop at 7 iteration ')
```
